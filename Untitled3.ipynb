Przeksztalcic sygnal sinusoidalny za pomocą DFT i wyświetlic wynik

```python
import numpy as np
import matplotlib.pyplot as plt
x = np.arange(0, np.pi*5, 0.1)
y= np.sin(x)
def dft(signal):
    N = len(signal)
    n =np.arange(0, N, 1)
    k = n.reshape(((N, 1)))
    e = np.exp(-2j*np.pi*((k*n))/N)
    return np.abs(np.dot(e, signal))
plt.title("DFT sinusa")
plt.xlabel("Częstotliwość")
plt.ylabel("Amplituda")
plt.plot(x[:int(len(x)/2)], dft(y)[:int(len(dft(y))/2)])
plt.show()
```

Porównac czasy wykonywania FFT i DFT

```python
import numpy as np
import matplotlib.pyplot as plt
import time
x = np.arange(0, np.pi*5, 0.1)
y= np.sin(x)+np.sin(16*x)+np.sin(8*x)+np.sin(4*x)+np.sin(2*x)+np.sin(12*x)+np.sin(6*x)
def dft(signal):
    N = len(signal)
    n =np.arange(0, N, 1)
    k = n.reshape(((N, 1)))
    e = np.exp(-2j*np.pi*((k*n))/N)
    return np.abs(np.dot(e, signal))
t1 = time.perf_counter()
dft(y)
t2 = time.perf_counter()
print("DFT: "+str(t2-t1))
t1 = time.perf_counter()
np.fft.fft(y)
t2 = time.perf_counter()
print("FFT: "+str(t2-t1))
```
